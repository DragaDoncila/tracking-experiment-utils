```python
import pandas as pd
import os

GT_DIV_ROOT = '/home/ddon0001/PhD/experiments/gt_div_constraint/'
GT_NO_DIV_ROOT = '/home/ddon0001/PhD/experiments/gt_no_div_constraint/'
DIV_ROOT = '/home/ddon0001/PhD/experiments/div_constraint/'
NO_DIV_ROOT = '/home/ddon0001/PhD/experiments/no_div_constraint/'

summary_df = pd.read_csv(os.path.join(GT_NO_DIV_ROOT, 'summary.csv'))
summary_df
```

```python
# load metrics for each sequence for each experiment
from collections import defaultdict
import json
res_dict = defaultdict(list)
for _, row in summary_df.iterrows():
    name = row['ds_name']
    with open(os.path.join(GT_DIV_ROOT, name, 'metrics.json')) as f:
        gt_old_metrics = json.load(f)
    with open(os.path.join(GT_NO_DIV_ROOT, name, 'metrics.json')) as f:
        gt_new_metrics = json.load(f)
    with open(os.path.join(DIV_ROOT, name, 'metrics.json')) as f:
        old_metrics = json.load(f)
    with open(os.path.join(NO_DIV_ROOT, name, 'metrics.json')) as f:
        new_metrics = json.load(f)
    res_dict['ds_name'].append(name)
    for key in gt_new_metrics:
        gt_old_key = 'gt_old_' + key
        gt_new_key = 'gt_new_' + key
        old_key = 'old_' + key
        new_key = 'new_' + key
        res_dict[gt_old_key].append(gt_old_metrics[key])
        res_dict[gt_new_key].append(gt_new_metrics[key])
        res_dict[old_key].append(old_metrics[key])
        res_dict[new_key].append(new_metrics[key])
res_df = pd.DataFrame(res_dict)
res_df
```

```python
res_df.to_csv(os.path.join(NO_DIV_ROOT, 'all_metrics.csv'), index=False)
```
