```python
import pandas as pd
import  os

SCALED_ROOT = '/home/ddon0001/PhD/experiments/scaled/'
KEYS = {
    'gt_old': 'gt_div_constraint',
    'old': 'div_constraint',
    'gt_new': 'gt_no_div_constraint',
    'new': 'no_div_constraint',
}
summary_df = pd.read_csv(os.path.join(SCALED_ROOT, KEYS['gt_new'], 'summary.csv'))
```

```python
from collections import defaultdict
import json
res_dict = defaultdict(list)
fake_dict = {}
for _, row in summary_df.iterrows():
    name = row['ds_name']
    res_dict['ds_name'].append(name)
    ds_metrics = {}
    for k, v in KEYS.items():
        current_path = os.path.join(SCALED_ROOT, v, name, 'metrics.json')
        # assign -1 metric for SIM datasets for running on ST
        if not os.path.exists(current_path):
            ds_metrics[k] = fake_dict
        else:
            with open(current_path) as f:
                metrics = json.load(f)
                ds_metrics[k] = metrics
    real_keys = []
    for sol_type, sol_metrics in ds_metrics.items():
        if not real_keys:
            real_keys = list(sol_metrics.keys())
        for metric_key in real_keys:
            new_key = f'{sol_type}_{metric_key}'
            res_dict[new_key].append(sol_metrics.get(metric_key, -1))
res_df = pd.DataFrame(res_dict)
res_df
```

```python
import re
res_columns = res_df.columns
bad_patterns = ['.*_fp_nodes', '.*_ns_nodes', 'gt_.*_DET', 'gt_.*_nodes', 'new_DET']
combined = "(" + ")|(".join(bad_patterns) + ")"
drop_cols = [col for col in res_columns if re.match(combined, col)]
filtered_df = res_df.drop(columns=drop_cols)
```
