```python
OLD = '/home/ddon0001/PhD/experiments/scaled/no_div_constraint/'
NEW = '/home/ddon0001/PhD/experiments/grid_again/best/'
USEFUL_KEYS = ['AOGM', 'fn_edges', 'ws_edges', 'fp_edges']
```

```python
import os
import pandas as pd
from collections import defaultdict
import json
datasets = os.listdir(NEW)
res_dict = defaultdict(list)
for ds_name in datasets:
    name = '_'.join(ds_name.split('_')[:-1])
    with open(os.path.join(OLD, name, 'metrics.json')) as f:
        old_metrics = json.load(f)
    with open(os.path.join(NEW, ds_name, 'metrics.json')) as f:
        new_metrics = json.load(f)
    res_dict['ds_name'].append(name)
    for key in USEFUL_KEYS:
        old_key = 'old_' + key
        new_key = 'new_' + key
        res_dict[old_key].append(old_metrics[key])
        res_dict[new_key].append(new_metrics[key])
res_df = pd.DataFrame(res_dict)
res_df
```

```python
penalty_root = '/home/ddon0001/PhD/experiments/grid_again/summaries/'
filenames = os.listdir(penalty_root)
cols = {}
for fn in filenames:
    summary = pd.read_csv(os.path.join(penalty_root, fn))
    cols['penalty'] = summary['penalty']
    name = '_'.join(fn.split('_')[:-1])
    cols[name] = summary['AOGM'] 
summary_df = pd.DataFrame(cols)
summary_df
```
